## Hva er fraktaler

Store norske leksikon definerer fraktaler som følgende: 

_"Fraktal, fellesnavn for strukturer og mønstre som er de samme uansett forstørrelse: de har egenlikhet og er uavhengig av skalering. Det er ingen dimensjon som er mer naturlig enn alle andre. Fraktaler er kalt naturens egen geometri."_

Utifra denne definisjonen har vi at fraktaler er spesielle mønstre som har uendelig med detaljer uansett forstørring eller skalering. En annen måte å beskrive fraktaler på, som definert av Benoit Mandelbrot, er som en delmengde i et kartesisk rom hvor dens Hausdorff-Besicovitch dimensjon er ekte større enn dens topologiske dimensjon. Fra et analytisk ståsted har man også at fraktaler ikke er deriverbare noe som helst sted.

Det vil si at de tre egenskapene som definerer en fraktal er:

- At Hausdorff-Besicovitch dimensjonen er ekte større enn den topologiske dimensjonen.
- En egenlikhet uansett forstørring av figuren.
- At figuren ikke er deriverbar for noe som helst sted.

Vi kan først gå dypere inn på det første punktet.

Hausdorff-Besicovitch dimensjonen er et mål på grovheten eller den fraktale dimensjonen i matematikken. For vanlige geometriske objekter vil Hausdorff-Besicovitch dimensjonen være lik den topologiske dimensjonen. For eksempel er den 0 for et punkt, 1 for en linje, 2 for et kvadrat og 3 for en kube, men det er også mulig å få Hausdorff dimensjoner som ikke er heltall (Gjelder for mange fraktaler). Det vil si at Hausdorff dimensjonen til et objekt er gitt som et reellt tall, $\mathbb{R}$, fremfor et naturlig tall, $\mathbb{N}$, som virker mer intuitivt.

For å løse for denne Hausdorff dimensjonen har man at en figur i $n$-te dimensjon som blir "rep-tiled"(Delt inn i biter som er lik orginalen) inn i biter skalert med en faktor $\frac{1}{r}$ ender opp med $r^{n}$ biter. Dersom vi tar Koch kurven under som eksempel har vi at vi kan dele den inn i fire kopier med en skaleringsfaktor på $\frac{1}{3}$. Nå kan vi finne dimensjonen til kurven ved å løse følgende:

$$3^{D} = 4$$

Som vi kan skrive om til:

$$D = \frac{\ln{4}}{\ln{3}}$$

Her har vi at $D \approx 1.2619$ (Større enn den topologiske dimensjonen, 1.)

![Koch kurven med tilfeldig orientering](https://upload.wikimedia.org/wikipedia/commons/2/23/Random_orientation_koch.png)
Koch kurven med tilfeldig orientering

Topologisk dimensjon er da de dimensjonene vi er kjent med: 0 for et punkt, 1 for en linje, 2 for et plan osv. Det er tilfeller hvor den topologiske dimensjonen ikke er like åpenbar, men det er ikke noe jeg kommer til å gå dypere inn i her.

Vi kan ta et eksempel hvor vi sjekker om Mandelbrotmengden er en fraktal.

![Mandelbrotmengden](https://upload.wikimedia.org/wikipedia/commons/d/d7/Mandel.png)
Her ser vi den første visualiseringen av Mandelbrotmengden som ble lagd.

Jeg har at Hausdorff dimensjonen til mandelbrotmengden er 2, dette er en verdi jeg har hentet fra nettet. Dersom vi ser på visualiseringen over ser vi at Mandelbrotmengden har en topologisk dimensjon av 2. Her har vi at Hausdorff dimensjonen og den topologiske dimensjonen er like, altså Mandelbrotmengden er faktisk ikke en fraktal. Det vi derimot kan se er at randen til Mandelbrotmengden har en topologisk dimensjon av 1 og fortsatt samme Hausdorff dimensjon på 2, altså randen til Mandelbrotmengden er en fraktal, men ikke selve figuren.

Egenlikheten er kanskje den egenskapen som er lettest å visualisere eller se. De fleste er vel bedre kjent med fraktalkunst enn fraktaler generelt, så man er kanskje kjent med konseptet om egenlikhet fra før av. For egenlikheten går ut på at man kan forstørre figuren i det uendelige og oppdage nye detaljer hvor det er de samme mønstrene som gjentar seg gang på gang.

Den siste egenskapen ved fraktaler er at de ikke er derivarbare i noe som helst sted. For å gjøre det mer konkret kan vi da ikke måle disse fraktalene med tradisjonelle metoder. Hvis vi skulle tatt lengden for eksempel, så vil vi aldri få noe eksakt svar, fordi figuren blir mer og mer detaljert desto mer vi forstørrer. Et eksempel på dette er det kjente problemet om å finne kystlengden til Storbrittania. Desto mer detaljert man måler, desto lengre blir kysten, og man vil alltid kunne være mer ekstakt, så det er umulig å finne et ekte mål på kystlengden.

Det er også mange matematikkere som mener det bør være flere kriterier for å kunne defineres som en fraktal:

- Fin og detaljert struktur på selv en liten skala.
- Irregularitet både lokalt og globalt.
- En enkel og gjerne rekursiv formel.

Med disse kriteriene på plass kan man få et bedre perspektiv over hva et fraktal er. For eksempel kunne man tro at en rett linje var en fraktal, ettersom den består kriteriet om egenlikhet, men den topologiske dimensjonen er lik dens Hausdorff dimensjon, og den er ganske lite detaljert uansett skalering.

Dersom man skulle plassert fraktalgeometri på kartet over matematikk ligger det naturligvis under måleteori. Det er matematikk hvor man kan systematisk plassere ulike verdier i ulike delmengder. Det er en generalisering av blant annet konseptene lengde, areal og volum. Fraktalgeometri er også sterkt knyttet opp mot blant annet annen geometri og topologi.

## Historie

Historien til fraktaler har gått fra først og fremst teoretiske studier til mer moderne bruk innenfor datagrafikk med flere nevneverdige navn som har påvirket feltet på sin egen måte. Den første bekjentskapen til noe som likner fraktaler som vi kjenner til idag kommer fra en vanlig afrikansk struktur for å bygge opp byer, hvor det var mindre strukturer hadde en tendens til å likne større strukturer. Det vil si at de for eksempel lagde en sirkulær landsby hvor alle husene/hyttene i landsbyen også var sirkulære. Det er kanskje ikke like kompleks som de fraktalene vi kjenner til idag, men man kan se tegn til den egenlikheten vi ser etter i fraktaler. 

Det er først på 1600-tallet vi begynner å se tegn til matten bak fraktaler når Gottfried Liebniz begynte å utforske den. Han er blant annet, sammen med Isac Newton, kjent for oppdagelsen av kalkulus. Han begynte å sette seg inn i ideen om rekursiv egenlikhet, men han gjorde en tabbe i å tro at kun en rett linje kunne ha denne egenskapen. Etter dette ble temaet lagt død en stund, og ofte refferert til som "monstre" innenfor matematikken. Det var først to århundre senere når Karl Weierstrass kom med en funskjon som, dersom grafet ned, hadde hatt den ikke veldig intuitive egenskapen om å være overalt kontinuerlig, men ingensted deriverbar. Slike funskjoner har i senere tid blitt kalt Weierstrassfunksjoner.

![Bilde av en Weierstrassfunksjon](https://upload.wikimedia.org/wikipedia/commons/6/60/WeierstrassFunction.svg)
En Funksjon Weierstrass fant innenfor intervallet $[-2, 2]$

Som vi kjenner fra egenskapene til fraktaler inngår egenskapen om å ikke være deriverbar i noe punkt. Vi kan også se på figuern at den har en egenlikhet, og med en Hausdorff-dimensjon på $1.5$ og en topologisk dimensjon på $1$ kan man nå bekrefte at dette er en fraktal. Slike funksjoner kan gis gjennom Fourier rekker, periodiske funksjoner lagd av harmonisk relaterte sinuser kombinert med en vektlagt summering, og den generelle formelen går slik:

$$f(x) = \sum_{n = 0}^{\infty}a^n\cos{(b^n\pi x)}$$

Hvor $0 < a < 1$, og $b = 2k + 1$, $k \in \mathbb{N}$
Vi har også at:

$$ab > 1 + \frac{3}{2}\pi$$

På grunn av disse begrensningene er $b = 7$ den minste verdien man kan ha av $b$.

Ikke lenger etter kom en av elevelene til Weierstrass, George Cantor, og introduserte et nytt konsept, Cantor sett. Cantor sett er nå kjent som fraktaler.

![Seks første iterasjonene av et Cantor sett](https://upload.wikimedia.org/wikipedia/commons/5/56/Cantor_set_in_seven_iterations.svg)
De seks første iterasjonene av et Cantor sett.

Et av de mest kjente Cantor settene er ternærsettet som avbildet over. Ideen er at man deler en linje i tre deler og midterste. Deretter sitter du igjen med to linjesegmenter hvor du gjør akkurat samme prosess. Dersom man itererer denne prosessen mange nok ganger vil man til slutt sitte igjen med en rekke punkter. Disse punktene har sine bruksområder, men det er ikke noe jeg kommer til å gå dypere inn i her.

Neste milepælen innenfor fraktalgeometri kom i 1904, når svenske Helge von Koch ga en mer geometrisk definisjon til fraktaler i opposisjon til Weierstrass sin analytiske definisjon og kom med kjente fraktaler vi idag kjenner, blant annet Koch snøflaket.

Et tiår senere konstruerte Waclaw Sierpinski sin kjente trekant og ett år senere Sierpinskiteppet.

![Utviklingen av Sierpinskitrekanten](https://upload.wikimedia.org/wikipedia/commons/0/05/Sierpinski_triangle_evolution.svg)
Utviklingen av Sierpinskitrekanten.

Ved 1918 kom to franske matematikere, Pierre Fatou og Gaston Julia, selv om de jobbet selvstendig fra hverandre, til den samme konklusjonen om fraktal oppførsel ved kartlegging av komplekse koordinater og iterative funksjoner.

Kort etter utviklet Felix Hausdorff dimensjons begrepet slik vi kjenner det, noe jeg har snakket om tidligere i rapporten, og dette ble veldig essentsielt i definisjonen av fraktaler, ettersom det ga muligheten for ikke-heltallige-dimensjoner. Hans dimensjons begrep baserte seg på grovheten i figurene.

Det var i lang tid vanskelig å jobbe visuelt med fraktaler ettersom alle visualiseringer måtte gjøres manuelt fremfor avansert datagrafikk vi har idag. Dette påvirket blant annet Juliamengdene som var vanskelige å visualisere i lang tid med mindre man bare gjorde det for et par iterasjoner. Dette endret seg derimot. I 1960, når Benoit Mandelbrot begynte å skrive om egenlikhet og samlet århundre med matematisk arbeid under begrepet "fraktaler", klarte han også å visualisere fraktaler gjennom illustrasjoner generert av datamaskiner. Disse visualiseringene, blant annet Mandelbrotmengden som er oppkallt etter han, har gitt oss det vi idag forbinder med fraktaler. Vakre, ofte rekursive, geometriske figurer.

### Fraktaltre
Et fraktaltre er et tre som deler seg i ledd som er lik seg selv, bare kortere. Det finnes flere måter å konstruere et fraktaltre på, men jeg tenker å konstruere et binært fraktaltre. Det vil si at hver gang treet deler seg, deler det seg i to. Viktige variabler man må ta hensyn til når man konstruerer fraktaltrær er vinkelen mellom grenene og lengden på grenene. Med mindre du velger noen kritiske lengder eller vinkler vil du ende opp med noe som likner et tre, men utseende kan variere stort. Av kjennetegnene til fraktaler jeg har snakket om tidligere vil det være mest intuitivt å se på egenlikheten i denne.

Tankegangen bak min implementasjon går ut på at jeg bruker Python og et bibliotek som heter pygame. Pygame er laget for å programmere spill i Python, men kan også brukes for andre visuelle modeller. I implementasjonen kommer jeg til å starte med en linje som går rett oppover med en bestemt lengde. Deretter tegner jeg to nye linjer fra endepunktet til den første linjen. Disse linjene kommer til å ha en lengde som er en bestemt skaleringsfaktor mindre enn den orinale linjen. Disse linjene kommer til å dele seg med en viss vinkel fra det den første linjen fulgte. Deretter repeterer jeg prosessen for de to nye linjene, også de fire neste osv. Det kommer til å være en rekursiv funksjon som kaller på seg selv frem til grenene er kortere enn en viss lengde og da stopper den. Ideelt skulle den gått på for evig, men ettersom det er begrenset med plass og tid er ikke det noe vits og det jeg har her er mer en bra nok for å skjønne ideen.

Obs: Når man kjører programmet er det best om man ikke trykker på noen taster og bare ser at programmet kjører i det eksterne vinduet, ellers vil programmet mistolke deg og stoppe opp. Da ryker kernelen som den kjører i og man må restarte den. Når man sier seg ferdig med å se på visualiseringen trykker men på x-en i høyre hjørne. For å kjøre programmet må du trykke deg inn i koderuten og trykke shift+enter. Dette gjelder for alle kodeblokker.

```python
#Importere moduler som jeg bruker i programmet
import pygame
import math
import time


#Fargeverdi og tykkelse
HVIT = (255, 255, 255)
tykkelse = 3

#Vinkelverdi, skaleringsverdi og minste grenlengde(Kan endres fritt på)
vinkelendring = math.pi/6
skaleringsverdi = 2/3
minstegrenlengde = 1

#Startpunktet
startx = 400
starty = 500

#Startverdier
startlengde = 150
startvinkel = math.pi/2


#Initialisere pygame
pygame.init()


#Eksternt vindu som kommer frem
skjerm = pygame.display.set_mode((800, 600))
skjerm.fill(HVIT)
pygame.display.set_caption("Fraktaltre")


#Oppdaterer skjermen
pygame.display.update()


#Rekursiv funksjon som lager treet
def tre(lengden, vinkel, startX, startY):
    global startvinkel
    global skaleringsverdi
    global skjerm
    global minstegrenlengde
    global tykkelse
    
    if lengden > minstegrenlengde:
        
        #Finner sluttpunktene til grenene
        sluttX = int(startX - lengden*math.cos(vinkel))
        sluttY = int(startY - lengden*math.sin(vinkel))
        
        #Tegner en gren
        pygame.draw.line(skjerm, (0, 255//lengden, 0), (startX, startY), (sluttX, sluttY), tykkelse)
        
        #Oppdaterer skjermen
        pygame.display.update()
        
        #Kaller funksjonen igjen for å tegne nye grener
        tre(lengden*skaleringsverdi, vinkel + vinkelendring, sluttX, sluttY)
        tre(lengden*skaleringsverdi, vinkel - vinkelendring, sluttX, sluttY)      

        
#Løkke som kjører frem til man har krysset ut vinduet
run = True
while run:
    
    #Når man trykker på x øverst i hjørnet sørger dette for at programmet lukker seg
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
            break
    
    
    #Gjør skjermen hvit
    skjerm.fill(HVIT)
    
    
    #Oppdaterer skjermen
    pygame.display.update()
    
    
    #Kaller funksjonen som lager treet
    tre(startlengde, startvinkel, startx, starty)


    #Oppdaterer skjermen
    pygame.display.update()


    #Holder figuren oppe i 5 sekunder
    time.sleep(5)
            
            
#Avslutter programmet
pygame.quit()
```

```python
#Importerer moduler jeg kommer til å bruke
import turtle
import math
import time


#Skilpadden
P = turtle.Turtle()


#Hvilket nivå den skal tegne, desto høyere, desto mer presist (kan endres fritt på)
i = 2


#vinkel gitt i grader
vinkel = 60


#Funksjon som definerer sierpinskikurven
def sierpinskikurve(order, lengde):
    global vinkel
    
    #Dersom orderen er et partall 
    if order % 2 == 0:
        kurve(order, lengde, vinkel)
    #Dersom den er et oddetall
    else:
        #Endrer litt på vinkelen om det er et oddetall
        P.lt(vinkel)
        kurve(order, lengde, -vinkel)
        

#Funksjon som definerer kurven
def kurve(order, lengde, vinkel):
    
    #Dersom orderen er 0 tegner den en linje
    if order == 0:
        P.forward(lengde)
    #Ellers blir en rekursiv prosess satt igang hvor den kaller på seg selv
    else:
        kurve(order - 1, lengde / 2, -vinkel)
        P.lt(vinkel)
        kurve(order - 1, lengde / 2, vinkel)
        P.lt(vinkel)
        kurve(order - 1, lengde / 2, -vinkel)

        
#Små modifikasjoner for synlighet og hastighet
P.speed("fastest") 
P.color("red")
P.pensize(2)


#Flytter skilpadden til startpunktet
P.pu()
P.goto(-50, 0)
P.pd()


#Tegner kurven
sierpinskikurve(i, 100)


#Lar programmet stå oppe i 5 sekunder
time.sleep(5)


#Vinduet lukker seg selv
turtle.bye()
```

### Barnsley Fern (Barnsleybregnen)

Barnsleybregnen har jeg ikke snakket så mye om tidligere, men det er etter min mening en av de peneste fraktalene. Det er en av de beste representasjonene av et egenlikt sett og er naturligvis lagd fra en rekursiv formel.

Barnsley bregnen bruker fire affine transformasjoner. Formelen for en affin transformasjon går som følger:

$$f(x, y) = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} + \begin{bmatrix}e \\ f \end{bmatrix}$$

Vi får da disse fire transformasjonene:

$$f_1(x, y) = \begin{bmatrix} 0.00 & 0.00 \\ 0.00 & 0.16 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix}$$

$$f_2(x, y) = \begin{bmatrix} 0.85 & 0.04 \\ -0.04 & 0.85 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} + \begin{bmatrix}0.00 \\ 1.60 \end{bmatrix}$$

$$f_3(x, y) = \begin{bmatrix} 0.20 & -0.26 \\ 0.23 & 0.22 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} + \begin{bmatrix}0.00 \\ 0.44 \end{bmatrix}$$

$$f_4(x, y) = \begin{bmatrix} -0.15 & 0.28 \\ 0.26 & 0.24 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} + \begin{bmatrix}0.00 \\ 0.44 \end{bmatrix}$$

Nå har vi verdiene for å kunne plotte bregnen, og så lenge vi tar utgangspunkt i verdiene over vil den samme bregneformen bli produsert. 

Det første punktet blir tegnet på et startpunkt $(x_0 = 0, y_0 = 0)$ og så blir nye punkter valgt tilfeldig gjennom å bruke en av de fire koordinattransformasjonene under:

$$f_1$$
$$x_{n + 1} = 0$$
$$y_{n + 1} = 0.16y_n$$

Denne koordinattransformasjonen er valgt 1% av tiden og kartlegger et hvilket som helst punkt til et punkt i det første linjesegmentet på basen av stilken. Denne delen av figuren er den første til å fullføres under produksjonen.

$$f_2$$
$$x_{n + 1} = 0.85x_n + 0.04y_n$$
$$y_{n + 1} = -0.04x_n + 0.85y_n + 1.6$$

Denne koordinattransformasjonen er valgt 85% av tiden og kartlegger et hvilket som helst punkt fra det store bladet til et av de mindre bladene.

$$f_3$$
$$x_{n + 1} = 0.2x_n - 0.26y_n$$
$$y_{n + 1} = 0.23x_n + 0.22y_n + 1.6$$

Denne koordinattransformasjonen er valgt 7% av tiden og kartlegger et hvilket som helst punkt i et av bladene til et tilsvarende punkt symmetrisk om stilken.

$$f_4$$
$$x_{n + 1} = -0.15x_n + 0.28y_n$$
$$y_{n + 1} = 0.26x_n + 0.24y_n + 0.44$$

Denne koordinattransformasjonen er valgt 7% av tiden og kartlegger et hvilket som helst punkt i bladene til et punkt på det samme bladet vridd om dens egen stilk.

Kort fortalt er det dette som skjer:
Den første koordinattransformasjonen tegner stilken, den andre genererer nye stilker på stilken for bladene. Den tredje genererer bladene nederst på den nederste til venstre. Den fjerde generer bladene på den nederste bladen til høyre. Og på grunn av rekursjon kan man garantere at hele bregnen blir en replikasjon av hver av bladene.

```python
import turtle
import random
import time

#Initialiserer skilpadden
P = turtle.Turtle()


#Endrer på hastighet og farge
P.speed("fastest")
P.color("green")
P.pensize(1)
P.penup()


#Startkoordinatene
x = 0
y = 0

#antall iterasjoner (kan endres på, men bør være stort for at det skal likne)
iterasjoner = 110

#Løkke som går like mange ganger som iterasjonen
for n in range(iterasjoner):
    
    
    #Flytter pennen
    P.goto(65 * x, 37 * y - 252)  
    P.pendown()
    
    #Plasserer et punkt der skilpadden er
    P.dot()
    P.penup()
    
    
    #Henter et tilfeldig tall mellom 0 og 1 og multipliserer det med 100
    r = random.random()
    r = r * 100
    
    
    #Setter x_n og y_n
    xn = x
    yn = y
    
    
    #If/else statement som baserer seg på tilfeldighetene jeg nevnte i forklaringen
    #Flytter så skilpadden til nye koordinater i neste omløp av løkken
    if r < 1:
        x = 0
        y = 0.16 * yn
    elif r < 86:
        x = 0.85 * xn + 0.04 * yn
        y = -0.04 * xn + 0.85 * yn + 1.6
    elif r < 93:
        x = 0.20 * xn - 0.26 * yn
        y = 0.23 * xn + 0.22 * yn + 1.6
    else:
        x = -0.15 * xn + 0.28 * yn
        y = 0.26 * xn + 0.24 * yn + 0.44

#Stopper opp programmet i 5 sekunder etter at det er ferdig med å tegne.
time.sleep(5)


#Lukker det eksterne vinduet
turtle.bye()
```

## Mandelbrotmengden

Til nå har jeg snakket mye om fraktaler, men jeg har ikke snakket så mye om komplekse tall. Den siste delen av denne rapporten derimot er rettet mot fraktaler hvor man utnytter komplekse tall. Det vil si i all hovedsak de to mest kjente: Mandelbrotmengden og Juliamengdene. Disse er mengder av tall på det komplekse planet som følger en viss trend kan man si.

Mandelbrotmengden defineres som mengden av komplekse tall $c$ for hvor funksjonen $f_c(x) = z^2 + c$ ikke divergerer når iterert fra $z = 0$.

Det vil si at vi må finne de komplekse tallene $c = a + bi$ hvor dette stemmer. Og for at det skal stemme har vi også at utrykket ikke skal divergere fra $z = 0$, hva vil det si?

Definisjonen av divergens går som følger: Følgen $\{a_n\}$ divergerer mot uendelig dersom det for ethvert tall $c \in \mathbb{R}$, finnes en $N \in \mathbb{N}$ slik at $a_n \geq c$ for alle $n \geq N$. I så fall skriver vi:

$$\lim_{n \to \infty} a_n = \infty$$

Tilsvarende sier vi at følgen $\{a_n\}$ divergerer mot minus uendelig dersom det for ethvert tall $c \in \mathbb{R}$, finnes en $N \in \mathbb{N}$ slik at $a_n \leq c$ for alle $n \geq N$. I så fall skriver vi:

$$\lim_{n \to \infty} a_n = -\infty$$

Da har vi altså at vi vil finne de verdiene, $c$, i det komplekse planet som ikke blåses opp mot uendelig noe sted dersom man fortsetter følgen til det n-te ledd. Det vil ikke nødvendigvis bety at verdiene konvergerer mot et spesifikt tall, det kan også innebære at det oscillerer mellom to eller flere verdier.

Nå sitter vi med en mulighet til å teste alle mulige komplekse tall, men det er veldig tidkrevende og strengt tatt umulig, så er det en måte å smale ned på hvilke tall vi kan velge mellom enda mer?

Heldigvis kan vi også si at verdien av $z_n$, til enhver tid, må ha en modulus, absoluttverdi, mindre enn 2. Det vil si at vi ikke kan ha $|c| > 2$ ettersom den vil divergere mot uendelig med engang. Vi har også at så fort en $z_n$ har beveget seg utenom sirkelen med radius 2, så vil den divergere, så det er ingen vits i å fortsette beregningene.

Vi kan ta et par eksempler:

$$c = 0$$
$$z_1 = 0^2 + 0 = 0$$
$$z_2 = 0^2 + 0 = 0$$
$$z_3 = 0^2 + 0 = 0$$
$$z_4 = 0^2 + 0 = 0$$

Her ser vi at den aldri vil bli noe større og divergerer altså ikke.

$$c = 1$$
$$z_1 = 0^2 + 1 = 1$$
$$z_2 = 1^2 + 1 = 2$$
$$z_3 = 2^2 + 1 = 5$$
$$z_4 = 5^2 + 1 = 26$$

Her derimot ser vi at $z$ blåser seg opp så fort den blir større enn 2, og vi har da at for $c = 1$ så divergerer utrykket. Altså $c = 1$ er ikke med i mandelbrotmengden.

Dersom man gjentar denne prosessen en rekke ganger innenfor begrensningen om at $|c| \leq 2$ og plotter de verdiene som ikke blåses opp i det komplekse vil man kanskje se et mønster. Jeg har tenkt å frembringe dette mønsteret gjennom metoden, men ettersom jeg ikke gidder å gjøre det for hånd bruker jeg heller programmering for å effektivisere prosessen.

Programmet mitt vil generere en rekke verdier for $c$ i formen $a + bi$ også iterere det gjennom formelen et bestemt antall ganger, og dersom den ikke blåses opp vil det plottes i en graf under programmet. Merk at desto flere iterasjoner man iterer gjennom formelen, desto mer nøyaktig tilnærming blir formen, men programmets kjøretid vil også bli lenger. Jeg kommer til å plotte fire grafer under programmet hvor x-aksen er den reelle aksen og y-aksen er den imaginære aksen. Forskjellen mellom de ulike grafene er hvor mange ganger de ulike verdiene for $c$ er iterert gjennom formelen.

```python
#Moduler som jeg importerer
from matplotlib import pyplot as plt
import cmath


def mandelbrotset(iterasjoner):

    #To lister som vil inneholde koordinatene til de ulike verdiene for c
    a = []    
    b = []
    
    
    #To løkker som genererer tallene i og j fra og med -200 til og med 200
    for i in range(-200, 200):
        for j in range(-200, 200):

            #Det komplekse tallet z som starter i origo
            z = complex(0, 0)
        
            #Den bestemte verdien for c for denne iterasjonen
            c = complex(i/100, j/100)

            #Ny løkke som itererer gjennom formelen
            for iterator in range(iterasjoner):
            
                #Dersom den blåses opp før antall iterasjoner er ferdig er det ingen grunn til å fortsette regningen
                if abs(z) > 2:
                    break
                
                #Beregningen for den nye z-en
                z = z**2 + c

            #Dersom z kom seg igjennom alle iterasjonene uten å blåse seg opp legges den til i listen over koordinatene til c
            if abs(z) <= 2:
                a.append(i/100)
                b.append(j/100)
    

    plt.scatter(a, b)
    plt.xlabel("Reell akse")
    plt.ylabel("Imaginær akse")
    plt.title("Mandelbrotmengde gitt {} iterasjoner gjennom formelen".format(iterasjoner))
    plt.grid()
    plt.show()
    
    
#Tilnærme mandelbrotmengden med 5 iterasjoner
mandelbrotset(5)
    
    
#Tilnærme mandelbrotmengden med 10 iterasjoner
mandelbrotset(10)
    
    
#Tilnærme mandelbrotmengden med 20 iterasjoner
mandelbrotset(20)
    
    
#Tilnærme mandelbrotmengden med 50 iterasjoner
mandelbrotset(50)
```

Dersom vi ser på hvor mange punkter $z$ oscillerer mellom for ulike verdier langs den reelle aksen, kan vi dra en sammenheng med et helt annet felt innenfor matematikken, nemlig kaosteori. Og da mer spesifikt logistisk vekst, men for å bedre forstå denne sammenhengen er det bedre å starte i andre enden, og til slutt resultere med mandelbrotmengden.

Det hele starter med en enkel, rekursiv formel:

$$x_{n + 1} = rx_n(1-x)$$

Hvor $x_{n + 1}$ er det neste leddet, $x_n$ er det nåværende leddet og $r$ er en konstant som fungerer som en vekstfaktor.

Man kan for eksempel se på denne formelen for logistisk vekst som en formel for populasjonsvekst. hvor vi har $x_n$ som dette års befolkning som en desimal fra og med 0 til og med 1, $x_{n + 1}$ som neste års befolkning som en desimal fra og med 0 til og med 1, og $r$ som vekstraten til befolkningen. Vi kan ved hjelp av dette grafe opp populasjonsutviklingen for kommende år gitt denne modellen for logistisk vekst:

```python
#Moduler som jeg importerer
from matplotlib import pyplot as plt
from numpy import zeros


#Antall iterasjoner
iterasjoner = 20


#Array som inneholder året n, og har en lengde lik iterasjoner + 1
n = zeros(iterasjoner + 1)

#Startåret
n_0 = 0

#Setter startverdien inn i arrayen
n[0] = n_0


#Array som inneholder populasjonen etter n antall år, og har en lengde lik iterasjoner + 1
x = zeros(iterasjoner + 1)

#Startverdien for x(Bare å endre på startverdien for x)
x_0 = 0.6

#Setter startverdien inn i arrayen
x[0] = x_0



#Vekstraten (Bare å endre på)
r = 2.6


#Løkke som itererer igjennom formelen
for i in range(iterasjoner):
    x[i+1] = x[i]*r*(1-x[i])
    n[i+1] = i + 1
    
    
#PLotte grafen jeg får
plt.grid()
plt.xlabel("år")
plt.ylabel("Populasjon")
plt.title("Populasjonsvekst over tid")
plt.plot(n, x)
plt.show()
```

Dersom man leker seg litt rundt ved å endre på vekstraten og startpopulasjonen vil man først merke to ting:

- Startpopulasjonen har ikke så mye å si, populasjonen vil stabilisere seg til "en" verdi over tid uansett.

- Vekstraten har mye å si for utviklingen av populasjonen.

Viktige verdier å se ved vekstraten er: hvis man endrer den til å være mindre enn 1 så vil populasjonen intuitivt nok dø ut over tid, hvis vekstraten er lik 1 så vil populasjonen forbli det samme gjennom hele perioden, og dersom man endrer vekstraten til å bli større enn 1 vil det stabilisere seg mot "en" verdi. Merk at jeg hele tiden har satt en i anførselstegn. Dette er fordi det viser seg at at den ikke stabiliserer seg mot en verdi for alle $r > 1$.Når man kommer til en viss verdi begynner den å oscillere mellom flere, også flere, også flere, også blir det totalt kaos. Det er her kaosteori kommer inn.

Kaosteori er en beskrivelse av visse ikke lineære dynamiske systemer som under visse vilkår oppfører seg som et fenomen kjent som kaos. Kaos går ut på at noe er veldig sensitivt til startvilkårene. Et underliggende konsept innenfor denne grennen er sommerfugleffekten, hvor man sier at små endringer kan føre til store konsekvenser. Et eksempel på dette kan være at en sommerfugl flagrer med vingene sine i Norge, og fører så til at man får orkan langs østkysten av Amerika. Det er ikke sagt at det skjer, men det sier noe om hvordan små handlinger kan få voldsomme konsekvenser. I dette tilfellet har vi at en liten endring i vekstfaktoren, kan føre til en kaotisk utvikling blant populasjonen. Dette kan visualiseres gjennom å grafe populasjonen etter en viss tid over vekstfaktoren i denne logistiske veksten vi har foretatt oss:

```python
#Moduler som jeg importerer
from matplotlib import pyplot as plt
from numpy import linspace
from random import randint


#Starverdien for populasjonen
x_0 = 0.5


#Antall år jeg skal måle populasjonen etter (tar ikke med slingringsmålet)
n = 70

#Lister som inneholder populasjonen etter en viss tid og vekstratene
X = []
R = []


#Løkke som endrer på r mellom intervallet [0, 4]
for r in linspace(0, 4, 10000):
    
    
    #Nåværende størrelse på populasjonen
    x = x_0
    
    #Kjører denne løkken n ganger pluss et slingringsmål for at grafen skal se bedre ut
    for i in range(n + randint(0, 10)):
        
        
        #Den rekursive formelen
        x = r*x*(1-x)
        
    
    #Setter de nåværende verdiene inn i listene
    R.append(r)
    X.append(x)
    
    
#PLotte grafen jeg får
plt.grid()
plt.xlabel("vekstrate")
plt.ylabel("Populasjon")
plt.title("Populasjonen over tid for ulike vekstfaktorer")
plt.scatter(R, X)
plt.show()
```

Her ser vi et litt kaotisk resultat etterhvert. I starten, frem til vekstraten når omtrent 3, så oppfører befolkningen seg slik man skulle forventet, men så ser den ut til å splitte seg avhengig av r. Den gjør egentlig ikke det, den oscillerer mellom to verdier, men siden jeg tok med det tilfeldige slingringsmålet, ser grafen mer komplett ut. Dersom man ser nøye kan man se at grafen splitter seg i to deler igjen ved enden på der den splittet seg i to første gangen, og det har seg slik at det gjør den uendelig mange ganger. Denne grafene er nemlig en fraktal med ganske tydelig egenliket, og (Jeg verken bekrefter eller beviser det her, men jeg vet at det stemmer) har en Hausdorff-dimensjon større enn dens topologiske dimensjon. Naturligvis er den også ikke deriverbar ved noe punkt.

Det viktige å få frem her er at vi ser den kaotiske utviklingen sommerfugleffekten satt i funskjon ved at vi bare endrer litt på vekstraten. ## Juliamengder

Juliamengder ligner litt på mandelbrotmengden. Begge har fraktalegenskaper, [begge lager sjuke mønstre ved forstørring som man kan sitte å se på på youtube i lang tid](https://www.youtube.com/watch?v=YOneAeBz8BQ), og slik jeg har forstått det til nå tar begge utgangspunkt i den samme formelen:

$$z_{n + 1} = z_n + c$$

Det er derimot en liten forskjell i hvordan man bruker formelen. Fra mandelbrotmengden starter vi alltid på $z_0 = 0$ og varierer $c$, mens i Juliamengder bestemmer vi oss for en konstant $c$ og tester formelen for ulike verdier av $z$. Vi gjør så sammebeslutning som ved mandelbrotmengden hvor vi avgjør om følgen vi får konvergerer mot en bestemt verdi, oscillerer mellom flere, eller divergerer.

En mer formel definisjon av Juliamengder går som følger:

Vi lar $f(z)$ være en ikke konstant holomorf funksjon fra den Riemanske sfæren over på seg selv. Slike $f(z)$ er helt presist  de ikke konstante rasjonale funksjonene, gitt ved $f(x) = \frac{p(z)}{q(z)}$, hvor $p(z)$ og $q(z)$ er komplekse polynomer. Anta at $p$ og $q$ har ingen like røtter og minst en har en grad større enn 1. Da er det et endelig antall åpne mengder $F_1, ..., F_r$ soom ikke endres av $f(z) og som er slik at:

1. Unionen av mengdene $F_i$ ligger tett opp til hverandre i planet.
2. $f(z)$ oppfører seg på regulær og lik måte på hvert av mengdene $F_i$.

For å forstå denne definisjonen her må vi først klargjøre noen begreper. 

Først og fremst hva en holomorf funksjon er. En holomorf funksjon er en kompleks og deriverbar funksjon, slik jeg har forstått det. Da har vi at den har komplekse variabler og er ved et hvert punkt deriverbar.

Deretter nevner definisjonen den Riemanske sfæren. Det viser seg å være en unik måte å visualisere det utvidede komplekse planet, slik at punktet uendelig et punkt på sfæren likesom et hvilken som helst annet kompleks tall. Ideen om den Riemanske sfæren gjør det altså enklere å tenke på deriverbarhet og kontinuitet i komplekse funksjoner.

Til slutt har vi, gitt to mengder A og B, at unionen til A og B er elementene i A, elementene i B og elementene i både A og B.

Da forteller definisjonen oss at vi har en kompleks funksjon $f(z)$ som inneholder et endelig antall åpne mengder, hvor mengdene ligger i nærheten av hverandre og funksjonen, $f(z)$, oppfører seg likegyldig ovenfor hver av mengdene.

Det som er verdt å merke seg ved denne definisjonen er at det fungerer slik som jeg forklarte før definisjonen, men funksjonen $f(z)$, trenger ikke å være på formen:

$$f(z) = z^2 + c$$

Selv om dette er den vanligste, og er den som har den største tilknyttelsen til Mandelbrotmengden.

En egenskap som nevnes hos Juliamengder er også at de er kaotiske. Som nevnt tidligere har vi da at en liten endring kan skape voldsomt store forandringer.

Nå har vi et utgangspunkt i hva julamengder går ut på. Og i motsetning til mandelbrotmengden har vi mange forskjellige former å utforske. Det er fordi, for hver eneste verdi for $c$, har vi en egen form eller geometrisk figur tilknyttet til den. For eksempel vil Juliamengden til $f(z) = z^2 + 0$ se annerledes ut enn Juliamengden til $f(z) = z^2  - 1$. Under her lager jeg et program som vil gi en tilnærming av hvordan mengden vil se ut for en valgfri verdi av $c$.

```python
#Moduler som jeg importerer
from matplotlib import pyplot as plt


#Funksjon som genererer en visuell tilnærming av en Juliamengde gitt en reell og imaginær verdi for c
def julia(a, b):
    
    
    #Definerer c som et komplekst tall
    c = complex(a, b)
    
    
    #To lister som kommer til å inneholde koordinatene til de ulike punktene som er i Juliamengden
    x = []
    y = []
    
    #To løkker som genererer tallene i og j fra og med -200 til og med 200
    for i in range(-200, 200):
        for j in range(-200, 200):    
            
            z = complex(i/100, j/100)
            
            for iterasjoner in range(100):
                
                #Dersom den blåses opp før antall iterasjoner er ferdig er det ingen grunn til å fortsette regningen
                if abs(z) >= 2:
                    break
                    
                
                #Itererer z opptil flere ganger for å se om den blåser seg opp.
                z = z**2 + c
                
                
            #Dersom z kom seg igjennom alle iterasjonene uten å blåse seg opp legges den til i listen over koordinatene til mengden
            if abs(z) < 2:
                x.append(i/100)
                y.append(j/100)   

    
    #Plotter punktene
    plt.scatter(x, y)
    plt.xlabel("Reell akse")
    plt.ylabel("Imaginær akse")
    plt.title("Juliamengden gitt c = {} + {}i".format(a, b))
    plt.grid()
    plt.show()

'''
Det er fullt mulig å endre på verdiene til c
og så se hvordan det vil se ut da

'''
#Juliamengden for c = 0
julia(0, 0)

#Juliamenden for c = -1
julia(-1, 0)

#Juliamengden for c = -1.037 + 0.17i
julia(-1.037, 0.17)

#Juliamengden for c = -0.52 + 0.57i
julia(-0.52, 0.57)

#Juliamengden for c = 0.295 + 0.55i
julia(0.295, 0.55)

#Julamengden for c = 1
julia(1, 0)
```

Her har jeg visualisert juliamengden for ulike verdier av c. Noen viktige trekk (hvor ikke alle blir belyst i mine nåværende resultater) er:

- for $c$ i origo i det komplekse planet har vi enhetssirkelen.

- verdier av $c$ som er i nærheten av hverandre, vil ha liknende former.

- Noen verdier for $c$ vil ikke ha noen stabile punkter hvor $z$ konvergerer helt.

Det første punktet om at c = 0 vil lage enhetssirkelen er rimelig intuitivt, ettersom vi da har:

$z_{n+1} = z_{n}^{2}$

Vi vet at 1 forblir 1 når det er kvadrert, verdier større enn 1 vokser når de blir kvadrert og verdier mindre enn en minsker når de blir kvadrert. Ettersom vi også bruker  det komplekse planet er det da modulusen vi ser på, og den fungerer som vanlige skalarer ved kvadrering (Vinkelen endres også, men det har ikke noe å si på resultatet).

Det andre punktet om at verdier av $c$ som er i nærheten av hverandre vil ha liknende former er litt mindre intuitivt, men dersom vi ser tilbake på en annen mengde vi har snakket om kan man se en sammenheng.

![Mandelbrotmengden, nei vent, mange små juliamengder?!](https://upload.wikimedia.org/wikipedia/commons/9/99/Julia_Mandelbrot_Relationship.png)
En Mandelbrotmengde bygd opp av mange små juliamengder

Her ser vi en tilnærming av mandelbrotmengden hvor hvert punkt er juliamengden for verdien av $c$ i det punktet. Det som er viktig å merke seg er hvordan juliamengdene som ligger ved siden av hverandre er nesten identiske. Dersom man skulle laget en film hvor man fulgte en linje med bilder av Juliamengder som ligger ved siden av hverandre, vil man merke en kontinuerlig flyt. Som om hele mengden forflytter seg. De viktigste formene å være kjent med er at:

- rundt origo vil Juliamengdene se ut som en sirkel
- I den større sirkelen under og over kardoiden har vi kaninmengdene, ettersom mengdene ser ut til å ha uendelig med kaninører.
- I den største sirkelen ved siden av kardoiden har vi avlange former som kan ligne en del på den orginale Mandelbrotmengden.
- Utover antennen langs den negative reelle aksen har vi veldig avlange former som nesten ser ut som linjestykker
- Og ved elefantdalen har vi det jeg liker å kalle for deformerte beyblades.

Nå som vi har fått forklart det andre punktet tror jeg ikke det blir så vanskelig å forklare det siste. Vi har nå etablert en sammenheng mellom de ulike Juliamengdene og Mandelbrotmengden. Og vi ser at for verdier av $c$ som befinner seg innenfor omkretsen til Mandelbrotmengden, så har vi geometriske former som ofte har et stort svart areal (som idenne betydning betyr at det er verdier som konvergerer fort, de andre fargene vil si at den vokser, men det handler om hvor fort den vokser). Hva skjer dersom vi velger en $c$ verdi som er utenom omkretsen til mandelbrotsettet? Det første man vil merke er at det er ingen steder som konvergerer helt. Det er derfor programmet mitt resulterer i en blank graf dersom man velger en verdi for $c$ utenfor Mandelbrotmengden, som $c = 1$. Vi vil derimot se hvordan ikke alle punkter divergerer i samme hastighet. Det er derfor det fortsatt er avbildet figurer på bildet jeg har lagt inn. En siste ting som skjer når vi tester for verdier av $c$ utenfor omkretsen til Mandelbrotmengden er at det ikke lenger er sammenhengende geometriske former (tar utgangspunkt i en kontinuerlig fargelegging av planet hvor hastigheten den divergerer på påvirker fargen, ikke en binær fargelegging som jeg har i programmene mine), det er heller det man kaller støv. Mange diskontinuerlige punkter strødd rundt planet, innenfor sirkelen med radius 2 da selvfølgelig. Når man kjenner til det siste argumentet jeg tok opp der, har man også en alternativ måte å definerer Mandelbrotmengden på: "Dersom juliamengden for det koordinatet i det komplekse planet er en sammenhengende geometrisk form, er det koordinatet inkludert i mandelbrotmengden, dersom det ikke er sammenhengende, men heller støv, er dette koordinatet utenfor mandelbrotmengden."

Nå har vi sett på Juliamengder og deres former og egenskaper. Vi har sett på hvordan de varierer og har en sammenheng med Mandelbrotmengden. Alt i alt er det skjønne former som på et vis er underlige og merkelige.
